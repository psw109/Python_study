# 그래프 축 레이블 설정

축 레이블 설정  


```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [2,3,5,10]

plt.plot(x, y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

## 축 레이블 여백 지정하기

```python
plt.plot(x, y)
plt.xlabel('x-axis', labelpad = 15)
plt.ylabel('y-axis', labelpad = 20)
```

## 폰트 설정하기

```python
plt.plot(x,y)
plt.xlabel('x', labelpad = 15, fontdict = {'family' : 'serif', 'color':'b', 'weight':'bold', 'size' : 14})
plt.ylabel('y', labelpad = 15, fontdict = {'family' : 'fantasy', 'color':'r', 'weight':'normal', 'size' : 10})
plt.show()
```

```python
# font인자 지정

font1 = {'family': 'serif',
         'color': 'b',
         'weight': 'bold',
         'size': 14
         }

font2 = {'family': 'fantasy',
         'color': 'deeppink',
         'weight': 'normal',
         'size': 'xx-large'
         }

plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis', labelpad=15, fontdict=font1, alpha = 0.5)
plt.ylabel('Y-Axis', labelpad=20, fontdict=font2)
plt.show()

#alpha : 텍스트 투명도 (0 ~ 1)
```

# 산점도 그리기

## 색상과 크기 지정하기

```python
np.random.seed(0)

n = 50
x = np.random.rand(n)
y = np.random.rand(n)

size = (30 * np.random.rand(n)) ** 2
colors = np.random.rand(n)

plt.scatter(x, y, s = size , c = colors)
plt.show()
```

## 투명도와 컬러맵 설정하기

```python
np.random.seed(0)

n = 50
x = np.random.rand(n)
y = np.random.rand(n)
size = (30 * np.random.rand(n))**2
colors = np.random.rand(n)

plt.scatter(x, y, s=size, c=colors, alpha=0.5, cmap='Spectral')
plt.colorbar()
plt.show()
```

# 3차원 산점도 그리기


```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

n = 100
xmin, xmax, ymin, ymax, zmin, zmax = 0, 20, 0, 20, 0, 50
cmin, cmax = 0,2

xs = np.array([(xmax - xmin) * np.random.random_sample() + xmin for i in range(n)])
ys = np.array([(ymax - ymin) * np.random.random_sample() + ymin for i in range(n)])
zs = np.array([(zmax - zmin) * np.random.random_sample() + zmin for i in range(n)])
color = np.array([(cmax - cmin) * np.random.random_sample() + cmin for i in range(n)])

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs, c=color, marker='o', s=15, cmap='Greens')

plt.show()
```

# 히스토그램 그리기

```python
import matplotlib.pyplot as plt
import numpy as np

weight = np.random.randint(50, 100, size = 30)
plt.hist(weight)
plt.show()
```

## 구간 개수(막대 개수) 지정하기

```python
np.random.seed(3)
weight = np.random.randint(50,100, size = 30)
plt.hist(weight, label = 'bins=10') # bins default 10
plt.hist(weight, bins = 30, label = 'bins=30')
plt.legend()
plt.show()
```

## 누적 히스토그램 그리기

```python
np.random.seed(3)
weight = np.random.randint(50,100, 30)

plt.hist(weight, cumulative = True, label = "cumulative")
plt.hist(weight, cumulative = False, label = "normal")
plt.legend(loc = 'upper left')
plt.show()
```

```python
np.random.seed(3)
weight = np.random.randint(50,100, 30)

plt.hist(weight, cumulative = True, label = "cumulative", density = True)
plt.hist(weight, cumulative = False, label = "normal", density = True)
plt.legend(loc = 'upper left')
plt.show()

#density = True를 해주면 면적의 합이 1이된다.
```
